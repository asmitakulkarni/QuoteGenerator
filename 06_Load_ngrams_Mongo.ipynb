```python
import pickle
from pymongo import MongoClient
from bson.binary import Binary
```

```python
client = MongoClient('localhost', 27017)
```

```python
db = client.character_models
collection = db.character
```

```python
all_characters = ['DUMBLEDORE', 'MCGONAGALL', 'HARRY', 'RON', 'HERMIONE', 'HAGRID', 'FRED', 'GEORGE', 'SNAPE', 
                  'PROFESSOR MCGONAGALL', 'TOM RIDDLE', 'SIRIUS BLACK', 'LUPIN', 'PROFESSOR LUPIN']
```

```python
all_characters
```

```python
def mongo_insert(character, model):
    model_id = collection.insert_one({'character': character, 'model': model}).inserted_id
```

```python
model_dict = {}
for character in all_characters:
    model_dict[character] = pickle.load(open(character + '.pkl', 'rb') )
    mongo_insert(character, Binary(open(character + '.pkl', 'rb').read()))
```
